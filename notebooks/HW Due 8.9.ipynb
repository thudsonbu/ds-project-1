Pandas need to be imported to complete this assignment. 

```python
import pandas as pd
```

We first import all tabs from the shipping1997workbook.

```python
shipping = pd.read_excel('shipping1997.xlsx', sheet_name = 'shipping')
states = pd.read_excel('shipping1997.xlsx', sheet_name = 'states')
commodities = pd.read_excel('shipping1997.xlsx', sheet_name = 'commodities')
distances = pd.read_excel('shipping1997.xlsx', sheet_name = 'distances')
```

Now that all tabs have been imported as dataframes, we can begin making the adjustments per the homework requirements.

We want to update the `shipping` dataframe to show state names along with the abbreviations.

```python
shipping_change_origin = pd.merge(shipping, states, how='outer', left_on='origin', right_on='Abbreviation')

shipping_change_origin = shipping_change_origin.drop(columns=['Abbreviation','GDP'])
shipping_change_origin = shipping_change_origin.rename(columns={'State':'Origin State Name'})

shipping_change_origin.head()
```

```python
shipping_final = pd.merge(shipping_change_origin, states, how='outer', left_on='dest', right_on='Abbreviation')

shipping_final = shipping_final.drop(columns=['Abbreviation','GDP'])
shipping_final = shipping_final.rename(columns={'State':'Destination State Name'})

shipping_without_abbreviations = shipping_final.drop(columns=['origin','dest'])

shipping_without_abbreviations.head()
```

Now that we have added `Origin State Name` and `Destination State Name` as  columns, we can continue

The column `adjacent` in the `distances` dataframe was shown as 1's nd 0's. Per the HW instructions, we had to change these values to the boolean datatype.

```python
distances['adjacent'] = distances['adjacent']. astype('bool')

distances.head()
```

The instructions wanted us to merge the `shipping` and `distances` dataframes. To do this, we used an outer join and merged on the `origin` and `dest` columns.

```python
shipping_distances = pd.merge(shipping_final, distances, how= 'outer', on=['origin','dest'])

shipping_distances.head()
```
